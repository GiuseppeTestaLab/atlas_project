# Testing scib metrics on cancer seacells integration with scvi implementation of scib

```python
import scanpy as sc
import pandas as pd
import scvi
from scib_metrics.benchmark import Benchmarker
```

```python
initDir = '/group/testa/Project/OvarianAtlas/atlas_project/raw_data/integration/metacells/cancer/'
```

```python
adata_2500 = sc.read(initDir + 'seacells_hdg_patients_batch_corr_scgen_tissuetreat_embeddings_2500.h5ad')
adata_list={'seacells_hdg_patients_batch_corr_scgen_tissuetreat_embeddings_2500.h5ad':'scGen_2500',
            'seacells_hdg_patients_batch_corr_scgen_tissuetreat_embeddings_HDG.h5ad':'scGen_HDG',
            'seacells_hdg_patients_batch_corr_scgen_tissuetreat_embeddings.h5ad':'scGen_all',
            'seacells_hdg_patients_batch_corr_scVI_tissuetreat_embeddings_2500.h5ad':'scVI_2500',
            'seacells_hdg_patients_batch_corr_scANVI_tissuetreat_embeddings_2500.h5ad':'scANVI_2500',
            'seacells_hdg_patients_batch_corr_scVI_tissuetreat_embeddings_HDG.h5ad':'scVI_HDG',
            'seacells_hdg_patients_batch_corr_scANVI_tissuetreat_embeddings_HDG.h5ad':'scANVI_HDG'}
adata_2500.obsm['X_umap_2500'] = adata_2500.obsm['X_umap']
embedding_obsm_keys = ['X_umap_2500']
for data in adata_list.keys():
    print('this is the path: '+ data)
    print('this is the nickname to be added in obsm: X_umap_' + adata_list[data])
    adata_temp = sc.read(initDir + data)
    adata_2500.obsm['X_umap_'+ adata_list[data]] = adata_temp.obsm['X_umap']
    del adata_temp
    embedding_obsm_keys.append('X_umap_'+ adata_list[data])
```

```python
bm = Benchmarker(
    adata_2500,
    batch_key="paper_ID",
    label_key="tissue-treatment",
    embedding_obsm_keys=embedding_obsm_keys,
    n_jobs=-1,
)
bm.benchmark()
```

```python
bm.plot_results_table(min_max_scale=False)
```

```python
bm.plot_results_table(min_max_scale=True)
```
