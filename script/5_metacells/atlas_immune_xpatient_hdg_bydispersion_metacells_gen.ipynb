```python
import numpy as np
import pandas as pd
import scanpy as sc
import SEACells
```

```python
adata = sc.read('/group/testa/Project/OvarianAtlas/Metacells_hdg_new/atlas_seacells_assignment_hdg_patients_immune.h5ad')
```

```python
adata.obs['SEACell_patient_tissue'] = adata.obs['SEACell'].astype('str') + '_' + adata.obs['paper_ID'].astype('str') + '_' + adata.obs['tissue'].astype('str')
```

```python
adata.obs.SEACell_patient_tissue.unique().shape
```

```python
adata = adata[~adata.obs.SEACell.isna(), :]
```

```python
adata_obs = adata.obs.copy()
```

```python
adata_obs = adata_obs.set_index('SEACell_patient_tissue')
adata_obs 
```

```python
adata_obs = adata_obs.loc[~adata_obs.index.duplicated(keep='first'), :]
```

```python
from SEACells.core import summarize_by_SEACell

ad = summarize_by_SEACell(adata, SEACells_label='SEACell_patient_tissue')
ad.layers['raw'] = ad.X
```

```python
ad.obs = adata_obs
```

```python
ad
```

```python
ad.obs
```

```python
ad.obs = ad.obs.drop(columns=['ID','patient_id', 'cell_type', 'cell_subtype', 'sample_ID', 'n_genes', 'n_genes_by_counts', 'total_counts', 'total_counts_mt', 'pct_counts_mt'])
```

```python
ad.obs
```

```python
ad.obs = ad.obs.drop(columns=['sample_name'])
```

```python
ad.obs['# Single Cells'] = adata.obs.groupby('SEACell_patient_tissue').count().tissue
ad.obs
```

```python
ad.write('/group/testa/Project/OvarianAtlas/Metacells_hdg_new/atlas_immune_seacells_hdg_patients.h5ad')
```
